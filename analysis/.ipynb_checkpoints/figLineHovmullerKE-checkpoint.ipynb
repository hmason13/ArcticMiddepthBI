```python
import matplotlib.pyplot as plt
import numpy as np
import netCDF4 as nc
%matplotlib inline
plt.rcParams['font.family'] = 'serif'
```

```python
# constants
secondsPerDay = 86400

# plotting params
deepIndex = 35

# io params
loc = f"/scratch/hcm7920/ambData/levelMean/"
conc = ["0", "100"]
saveloc = "/home/hcm7920/experiments/arcticMiddepthBI/plots/qdPlots/"                                       
```

```python
# load data

levs = np.load(loc+"../grid/gridLevs.npy")

timeDiags = np.load(loc+"../grid/gridTimeDiags.npy")
keLevMax = np.load(loc+"levelMaxKE.npy")
keLevMean = np.load(loc+"levelMeanKE.npy")

channelMeanU  = np.load(loc+"../channelMean/channelMeanU.npy")
channelMeanV  = np.load(loc+"../channelMean/channelMeanV.npy")
channelMeanKE = np.load(loc+"../channelMean/channelMeanKE.npy")
```

```python
# make plot

for i in range(len(conc)):
    plt.figure(figsize=(8,4))
    plt.contourf(timeDiags,levs[:deepIndex],keLevMean[:,:deepIndex,i].T,
                 15,cmap="magma")
    plt.xlabel("Time (days)")
    plt.ylabel("Depth (m)")
    plt.title(r"Level average kinetic energy ($\mathrm{m}^2/\mathrm{s}^2$), "                
              +f"{conc[i]}% cice, diagnostic")
    plt.colorbar()
    plt.savefig(saveloc+f"levelAverageKE{conc[i]}HovDiagnostic.png",
                bbox_inches="tight", pad_inches=0)
    
plt.clf()
plt.close('all')
```

```python
# make another plot
for i in range(len(conc)):
    levsPlotted = 7
    colors = plt.cm.viridis(np.linspace(0,0.9,levsPlotted))
    plt.figure(figsize=(8,4))
    for index in range(levsPlotted):
        plt.plot(timeDiags,keLevMean[:,index*4,i],color=colors[index],
                 label=f"{levs[index*4]:.0f} meters",linewidth=3)
    plt.xlabel("Time (days)")
    plt.ylabel("KE ($\mathrm{m}^2/\mathrm{s}^2$)")
    plt.title(f"KE @ selected levels, {conc[i]}% cice, diagnostic")
    plt.legend()
    
plt.clf()
plt.close('all')
```

```python
# combined line plot comparing two runs
levsPlotted = 4
interval = 5
colors = plt.cm.viridis(np.linspace(0,0.85,levsPlotted))
plt.figure(figsize=(8,4))
for index in range(levsPlotted):
    plt.plot(timeDiags,keLevMean[:,index*interval,0],color=colors[index],
             label=f"{levs[index*interval]:.0f} meters", 
             linewidth=3, linestyle="solid")
    plt.plot(timeDiags,keLevMean[:,index*interval,1],color=colors[index],
             label=f"",#{levs[index*interval]:.0f} meters", 
             linewidth=3, linestyle="dashed")
plt.xlabel("Time (days)")
plt.ylabel("KE ($\mathrm{m}^2/\mathrm{s}^2$)")
plt.title(f"Level avg KE, {conc[0]}% solid, {conc[1]}% dashed")
plt.legend()

plt.clf()
plt.close('all')
```

```python
# make hov plot for maximum
for i in range(len(conc)):
    plt.figure(figsize=(8,4))
    plt.contourf(timeDiags,levs[:deepIndex],keLevMax[:,:deepIndex,i].T,
                 15,cmap="magma")
    plt.xlabel("Time (days)")
    plt.ylabel("Depth (m)")
    plt.title(r"Level max kinetic energy ($\mathrm{m}^2/\mathrm{s}^2$), "
              +f"{conc[i]}% cice, diagnostic")
    plt.colorbar()
    
plt.clf()
plt.close('all')
    
```

```python


# make another line plot for maximum
for i in range(len(conc)):
    levsPlotted = 7
    colors = plt.cm.viridis(np.linspace(0,0.9,levsPlotted))
    plt.figure(figsize=(12,6))
    for index in range(levsPlotted):
        plt.plot(timeDiags,np.sqrt(keLevMax[:,index*5,i])*100,
                 color=colors[index],label=f"{levs[index*5]:.0f} meters")
    plt.xlabel("Time (days)")
    plt.ylabel("Velocity ($\mathrm{cm}/\mathrm{s}$)")
    plt.title(f"Max vel (via KE), selected levels, {conc[i]}% cice, diags")
    plt.legend()
    plt.savefig(saveloc+f"levelMaxVel{conc[i]}LineDiags.png",
                bbox_inches="tight", pad_inches=0)
    
plt.clf()
plt.close('all')
```

```python
# make line plot vs depth
deepIndex = 50

plt.plot(100*np.sqrt(keLevMean[:,:deepIndex,0].mean(0)),
        levs[:deepIndex], color="black",
        label="Ice free")
plt.plot(100*np.sqrt(keLevMean[:,:deepIndex,1].mean(0)),
        levs[:deepIndex], color="blue",
        label="Maximum ice")
plt.ylabel("Depth (m)")
plt.xlabel(r"RMS speed $(\mathrm{cm}\,\mathrm{s}^{-1})$")
plt.legend()
```
