```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# params
loc = "/scratch/hcm7920/ambData/"
conc = ["0", "40", "80", "100"]
deepIndex = 50
xRange = np.linspace(50,450,200)
```

```python
# load data

uCM = np.load("/scratch/hcm7920/ambData/channelMeanUSpinup.npy")[:,:,1,:]
wCM = np.load("/scratch/hcm7920/ambData/channelMeanWSpinup.npy")[:,:,1,:]
sCM = np.load("/scratch/hcm7920/ambData/channelMeanSaltSpinup.npy")[:,:,1,:]

usBar = np.load("/scratch/hcm7920/ambData/channelMeanUSBarSpinup.npy")
wsBar = np.load("/scratch/hcm7920/ambData/channelMeanWSBarSpinup.npy")

levs = np.load(loc+"gridLevs.npy")
```

```python
vmn = -1.9e-5
vmx =  1.9e-5
for i in range(len(conc)):
    plt.figure(figsize=(12,4))
    data=wsBar[:,:deepIndex,i].T-(wCM[25:-25,:deepIndex,i]*sCM[25:-25,:deepIndex,i]).T
    myMax=np.max(np.abs(data))
    contours = np.linspace(-myMax,myMax,26)
    plt.contourf(xRange,levs[:deepIndex],data,
                contours,cmap='seismic')#,vmax=myMax,vmin=-myMax)
    plt.colorbar()
    plt.title("$\overline{w's'}$"+f", {conc[i]} concentration")
    plt.xlabel("Across channel distance (km)")
    plt.ylabel("Depth (meters)")
```

```python
for i in range(len(conc)):
    plt.figure(figsize=(12,4))
    data=usBar[:,1:deepIndex,i].T-(uCM[:,1:deepIndex,i]*sCM[:,1:deepIndex,i]).T
    myMax=np.max(np.abs(data))
    contours = np.linspace(-myMax,myMax,26)
    plt.contourf(xRange,levs[1:deepIndex],data,
                contours,cmap='seismic')
    plt.colorbar()
    plt.title("$\overline{u's'}$"+f", {conc[i]} concentration")
    plt.xlabel("Across channel distance (km)")
    plt.ylabel("Depth (meters)")
```
